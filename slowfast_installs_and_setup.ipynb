```python
download_data_bash = '''DATA_DIR="data/ava/videos"

if [[ ! -d "${DATA_DIR}" ]]; then
  echo "${DATA_DIR} doesn't exist. Creating it.";
  mkdir -p ${DATA_DIR}
fi

wget https://s3.amazonaws.com/ava-dataset/annotations/ava_file_names_trainval_v2.1.txt

head -n 20 ava_file_names_trainval_v2.1.txt | while read line; do
  wget https://s3.amazonaws.com/ava-dataset/trainval/$line -P ${DATA_DIR}
done

IN_DATA_DIR="data/ava/videos"
OUT_DATA_DIR="data/ava/videos_15min"

if [[ ! -d "${OUT_DATA_DIR}" ]]; then
  echo "${OUT_DATA_DIR} doesn't exist. Creating it.";
  mkdir -p ${OUT_DATA_DIR}
fi

for video in $(ls -A1 -U ${IN_DATA_DIR}/*)
do
  out_name="${OUT_DATA_DIR}/${video##*/}"
  if [ ! -f "${out_name}" ]; then
    ffmpeg -ss 900 -t 901 -i "${video}" "${out_name}"
  fi
done


IN_DATA_DIR="data/ava/videos_15min"
OUT_DATA_DIR="data/ava/frames"

if [[ ! -d "${OUT_DATA_DIR}" ]]; then
  echo "${OUT_DATA_DIR} doesn't exist. Creating it.";
  mkdir -p ${OUT_DATA_DIR}
fi

for video in $(ls -A1 -U ${IN_DATA_DIR}/*)
do
  video_name=${video##*/}

  if [[ $video_name = *".webm" ]]; then
    video_name=${video_name::-5}
  else
    video_name=${video_name::-4}
  fi

  out_video_dir=${OUT_DATA_DIR}/${video_name}/
  mkdir -p "${out_video_dir}"

  out_name="${out_video_dir}/${video_name}_%06d.jpg"

  ffmpeg -i "${video}" -r 30 -q:v 1 "${out_name}"
done

'''

with open('setup_ava_dataset.sh', 'w') as f:
  f.write(download_data_bash)
```

```python
import os
import pandas as pd

# === Configuration ===
annotations_path = "data/ava/ava_annotations"
frames_path = "data/ava/frames"

# Output files (same folder as input annotations)
filtered_files = {
    "ava_train_v2.2.csv": "ava_train_tiny_v2.2.csv",
    "ava_val_v2.2.csv": "ava_val_tiny_v2.2.csv",
    "person_box_67091280_iou90/ava_detection_train_boxes_and_labels_include_negative_v2.2.csv": "ava_det_train_tiny_v2.2.csv",
    "person_box_67091280_iou90/ava_detection_val_boxes_and_labels.csv": "ava_det_val_tiny_v2.2.csv",
    "train.csv" : "train_tiny.csv",
    "val.csv" : "val_tiny.csv"
}

# === Step 1: Get available video_ids from frame folders ===
available_video_ids = set(os.listdir(frames_path))
print(f"Found {len(available_video_ids)} video IDs from frames directory.")

# === Step 2: Function to filter each CSV ===
def filter_ava_csv(file_path, output_path, video_ids):
    df = pd.read_csv(file_path, header=None)
    filtered_df = df[df[0].isin(video_ids)]
    filtered_df.to_csv(output_path, index=False, header=False)
    print(f"✔ Saved {len(filtered_df)} rows to {output_path}")


def filter_frame_list_ava_csv(file_path, output_path, video_ids):
    df = pd.read_csv(file_path, sep=r'\s+', engine='python')
    filtered_df = df[df["original_vido_id"].isin(video_ids)]
    filtered_df["labels"] = '""'
    filtered_df.to_csv(output_path, sep=" ", index=False, quoting=3)
    print(f"✔ Saved {len(filtered_df)} rows to {output_path}")


# === Step 3: Run filtering ===
for src_file, dest_file in filtered_files.items():
    full_src = os.path.join(annotations_path, src_file)
    full_dest = os.path.join(annotations_path, dest_file)

    if not os.path.exists(full_src):
        print(f"⚠ File not found: {full_src}")
        continue

    os.makedirs(os.path.dirname(full_dest), exist_ok=True)
    if src_file in ["train.csv", "val.csv"]:
        filter_frame_list_ava_csv(full_src, full_dest, available_video_ids)
    else:
        filter_ava_csv(full_src, full_dest, available_video_ids)
```

```python
keep_files = ["ava_train_tiny_v2.2.csv", "ava_val_tiny_v2.2.csv", "train_tiny.csv", "val_tiny.csv", "ava_action_list_v2.2_for_activitynet_2019.pbtxt"]
src_folder = "data/ava/ava_annotations"
dest_folder = "data/ava/annotations"
if not os.path.exists(dest_folder):
    os.makedirs(dest_folder)

for file in keep_files:
    src_file = os.path.join(src_folder, file)
    dest_file = os.path.join(dest_folder, file)
    shutil.copy(src_file, dest_file)
```
